# SPOTIFY TRENDS

In this notebook, we will briefly explore the Spotify Top 100 Songs from 2010-2019 Dataset from kaggle, found at the link below. 

https://www.kaggle.com/datasets/muhmores/spotify-top-100-songs-of-20152019?resource=download

Let us import this dataset as well as some of the necessary packages that we will use. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Spotify 2010 - 2019 Top 100 Songs.xlsx")
data.head()
```

```python
#Check data types and null value counts
data.info()
```

We see there is no missing data. This makes our job a little easier. Let's take a look at the value counts for each genre. 

```python
data['top genre'].value_counts()
```

We see there are 132 different genres listed (they get pretty specific). The top genre is dance pop, by far, followed by pop, atl hip hop, art pop, and boy band. 

Let's compare the popularity of the songs to various different variables such as danceability, how positive the mood of the song is, and duration. We want to learn more about what makes a song more popular. 

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='pop',y='dnce',data=data)
plt.title("Popularity vs Danceability")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='pop',y='val',data=data)
plt.title("Popularity vs Positivity")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x='pop',y='dur',data=data)
plt.title("Popularity vs Duration")
plt.show()
```
